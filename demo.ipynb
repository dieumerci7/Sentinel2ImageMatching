A small demo of how my code works.

```python
import matplotlib.pyplot as plt
import cv2
```

```python
from algorithm import match
```

We shall draw at most 10 matches for the purpose of this demo.

```python
N_MATCHES = 10
```

The images used are from Kaggle dataset, resized to 7000x7000 so that I could load them to GitHub (the code would work just as well with full-sized images).
the resizing code is as follows:

resized_image = src.read(
            window=Window(0, 0, new_width, new_height),
            out_shape=(src.count, new_height, new_width),
            resampling=Resampling.bilinear
        )

```python
image_path1 = './images/T36UYA_20160212T084052_B02.jp2'
image_path2 = './images/T36UYA_20160212T084052_B03.jp2'
```

```python
img_matches = match(image_path1, image_path2, n_matches=N_MATCHES)
```

Although I display the matches here using matplotlib, I recommend viewing them in './matches/result_different_bands.jpg'. The images are very dark, therefore the lines might be poorly visible in working environment.

```python
cv2.imwrite('./matches/result_different_bands.jpg', img_matches)
```

```python
# Display the result using matplotlib
plt.imshow(img_matches)
plt.axis('off')
plt.show()
```

It seems that the matching works rather well on different bands of the same image. It makes sense for corresponding keypoints to be on the same place in both bands. Still, I would stress that the method used does not allow to evaluate the performance exactly.
